```python
import time
import math
space=1009  ### es un numero primo 
n=150000000
```

```python
def formula(x,y):
    return x*x+y*y
```

```python
##Lagged Fibonacci generator
j = 3
k = 7
s = [int(time.time()), int(time.time()), int(time.time()), int(time.time()), int(time.time()), int(time.time()), int(time.time())]
def LFG():
    global s
    for i in range(0,len(s)):
        if i == 0:
            out = (s[j-1] + s[k-1]) %space  
        elif 0 < i < 6:
            s[i] = s[i+1] 
        else:
            s[i] = out
            return out
```

```python
from mpl_toolkits import mplot3d
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
def f(x, y):
    return x*x+y*y

x = np.linspace(-2,2, 30)
y = np.linspace(-2,2, 30)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 50, cmap='binary')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z');
```

```python
lado=4/space
cubito=lado*lado*lado
volumen=0
con=0
TIntentos=10
##lim=int(n/TIntentos)
print("Tepox se la come")
suma=0
for k in range(0,1):
    volumen=0
    con=0
    for i in range (0,n):
        x= LFG()
        x*=lado
        y= LFG()
        y*=lado
        z= LFG()
        z*=lado
        zr=formula(x,y)
        if z<=zr:
            volumen+=cubito
            con+=1
    suma+=volumen
    print("Aciertos=",con,"%=",con/n)
    print("Volumen Bajo la Curva=",volumen)

print("Lado=",lado,"volumen cubito=",cubito)
print("contador=",con,"n=",n)
print("Aciertos=",con,"%=",con/n)
print("Volumen Bajo la Curva=",volumen)
print("Promedio:",suma/TIntentos)
```
