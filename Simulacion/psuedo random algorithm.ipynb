```python
##Multiplicative Congruential Generator
def MCG(semilla,mod):
    mul=7873
    return (semilla*mul)%mod
```

```python
##Lagged Fibonacci generator
j = 3
k = 7
s = [int(time.time()), int(time.time()), int(time.time()), int(time.time()), int(time.time()), int(time.time()), int(time.time())]
def LFG():
    global s
    for i in range(0,len(s)):
        if i == 0:
            out = (s[j-1] + s[k-1]) %space  
        elif 0 < i < 6:
            s[i] = s[i+1] 
        else:
            s[i] = out
            return out
```

```python
 import math 
def dis(cnx,cny,px,py):
    dsx=abs(cnx-px)
    dsx*=dsx
    dsy=abs(cny-py)
    dsy*=dsy
    return math.sqrt(dsx+dsy)
```

```python
con=0
pcy=8000
pcx=4000
rd=2000
cx=[] 
cy=[]
nx=[] 
ny=[]
for i in range (0,n):
    val=dis(pcx,pcy,x[i],y[i])
    if val<=rd:
        con+=1
        cx.insert(len(cx), x[i] )
        cy.insert(len(cy), y[i] )
    else:
        nx.insert(len(nx), x[i] )
        ny.insert(len(ny), y[i] )
```

```python
import matplotlib.pyplot as plt
plt.plot(nx,ny,'r,')
plt.plot(cx,cy,'b,')
plt.show()
print("Posiciones dentro del area: ",con)
print("La probabilidad de acertar al centro es de",(con/n))
```
