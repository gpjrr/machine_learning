```python
import time
import math
space=1009  ### es un numero primo 
n=30000000
```

```python
##Multiplicative Congruential Generator
def formula(x,y):
    return x*x+y*y
def MCG(semilla,mod):
    mul=7873
    return (semilla*mul)%mod
```

```python
from mpl_toolkits import mplot3d
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
def f(x, y):
    return x*x+y*y

x = np.linspace(-2, 2, 20)
y = np.linspace(-2,2, 20)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 50, cmap='binary')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z');
```

```python
lado=4/space
cubito=lado*lado*lado
volumen=0
con=0
TIntentos=10
##lim=int(n/TIntentos)
suma=0
for k in range(0,2):
    volumen=0
    con=0
    for i in range (0,n):
        nuevo=int(time.time())
        nuevo=MCG(nuevo,space)
        x=nuevo
        x*=lado
        nuevo=MCG(nuevo,space)
        y=nuevo
        y*=lado
        nuevo=MCG(nuevo,space)
        z=nuevo
        z*=lado
        zr=formula(x,y)
        if z<=zr:
            volumen+=cubito
            con+=1
    suma+=volumen
    print("Aciertos=",con,"%=",con/n)
    print("Volumen Bajo la Curva=",volumen)

print("Lado=",lado,"volumen cubito=",cubito)
print("contador=",con,"n=",n)
print("Aciertos=",con,"%=",con/n)
print("Volumen Bajo la Curva=",volumen)
print("Promedio:",suma/TIntentos)
```
