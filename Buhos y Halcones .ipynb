```python
import pandas as pd
data1 = {'buhos': [151.0],
        'halcones': [199.0],}
# Convert the dictionary into DataFrame
df1 = pd.DataFrame(data1)
  
for i in  range(1,30):
    o=df1['buhos'][i-1]
    h=df1['halcones'][i-1]
    df1.loc[i]=[1.2*o-0.001*o*h ,1.3*h-0.002*o*h ]
# Observe the result
df1
```

```python
import matplotlib.pyplot as plt


ax = plt.gca()


plt.plot(df1['buhos'])
plt.plot(df1['halcones'])
plt.ylabel('total de animales')
plt.xlabel('dias')
##plt.plot(df1['halcones'],kind='line',x='dias',y='num_buhos')
plt.show()
```
